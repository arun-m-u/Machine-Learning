```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
datasets.load_breast_cancer()
```

```python
breast_cancer = datasets.load_breast_cancer()
```

```python
print(breast_cancer.DESCR)
```

## Converting to dataframe

```python
breast_cancer_data_df = pd.DataFrame(breast_cancer.data,columns = breast_cancer.feature_names)
breast_cancer_data_df
```

```python
breast_cancer_data_df["Target"] = breast_cancer.target
```

```python
breast_cancer_data_df
```

## Data understanding

```python
breast_cancer_data_df.shape
```

```python
breast_cancer_data_df.dtypes
```

## Model Building

```python
X = breast_cancer_data_df.drop(labels = "Target",axis = 1)
y = breast_cancer_data_df["Target"]
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scaled_x = pd.DataFrame(std_scaler.fit_transform(X = X),columns = breast_cancer.feature_names)
scaled_x
```

```python
X_train,X_test,y_train,y_test = train_test_split(scaled_x,y,test_size = 0.20,random_state = 123,shuffle = True,stratify=y)
```

```python
y_count = y.value_counts()
print(y_count)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.pie(y_count,labels = ["benign","malignant"],autopct = "%1.0f%%")
plt.show()
```

```python
y_train_count = y_train.value_counts()
```

```python
y_test_count = y_test.value_counts()
```

```python
plt.pie(y_train_count,labels = ["benign","malignant"],autopct = "%1.0f%%")
plt.show()
```

```python
plt.pie(y_test_count,labels = ["benign","malignant"],autopct = "%1.0f%%")
plt.show()
```

## Model Training

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier( criterion='gini', max_depth=3 , class_weight={0:2,1:1})
```

```python
dt_model.fit(X_train,y_train)
```

## Model Testing and Model Evaluation

## for training data

```python
y_predict_train = dt_model.predict(X_train)
y_predict_train
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_train,y_predict_train))
```

```python
print(confusion_matrix(y_train,y_predict_train))
```

## for test data

```python
y_predict_test = dt_model.predict(X_test)
y_predict_test
```

```python
print(classification_report(y_test,y_predict_test))
```

```python
print(confusion_matrix(y_test,y_predict_test))
```
